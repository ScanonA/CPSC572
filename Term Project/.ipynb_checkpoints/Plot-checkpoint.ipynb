```python
G = nx.Graph()
import csv

edgelist = []

with open('compiledEdgeList.csv') as f:
    reader = csv.reader(f, delimiter=',', quotechar='|')
    line = 0
    for row in reader:
        if line == 0:
            #print("Headers")
            line += 1
        else:
            edgelist.append([row[0],row[1]])
            line += 1
G.add_edges_from(edgelist)
```

```python
N = len(G)
L = G.size()
degrees = [G.degree(node) for node in G]
kmin = min(degrees)
kmax = max(degrees)

print("Number of nodes: ", N)
print("Number of edges: ", L)
print()
print("Average degree: ", 2*L/N)
print("Average degree (alternate calculation)", np.mean(degrees))
print()
print("Minimum degree: ", kmin)
print("Maximum degree: ", kmax)
```

```python
# from networkx.algorithms import community
#https://graphsandnetworks.com/community-detection-using-networkx/
import networkx as nx
import networkx.algorithms.community as nxcom

from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams.update(plt.rcParamsDefault)
plt.rcParams.update({'figure.figsize': (15, 10)})
# get reproducible results
import random
from numpy import random as nprand
random.seed(123)
nprand.seed(123)

G2 = nx.read_edgelist('./compiledEdgeList.csv', comments='#',
                     create_using=nx.Graph(), 
                     delimiter=',', 
                     nodetype=str, 
                     encoding='utf-8')
# Find the communities
communities = sorted(nxcom.greedy_modularity_communities(G2), key=len, reverse=True)
# Count the communities
print(f"The network has {len(communities)} communities.")
```

```python
def set_node_community(G, communities):
        '''Add community to node attributes'''
        for c, v_c in enumerate(communities):
            for v in v_c:
                # Add 1 to save 0 for external edges
                G.nodes[v]['community'] = c + 1
def set_edge_community(G):
    '''Find internal edges and add their community to their attributes'''
    for v, w, in G.edges:
        if G.nodes[v]['community'] == G.nodes[w]['community']:
            # Internal edge, mark with community
            G.edges[v, w]['community'] = G.nodes[v]['community']
        else:
            # External edge, mark as 0
            G.edges[v, w]['community'] = 0
def get_color(i, r_off=1, g_off=1, b_off=1):
    '''Assign a color to a vertex.'''
    r0, g0, b0 = 0, 0, 0
    n = 16
    low, high = 0.1, 0.9
    span = high - low
    r = low + span * (((i + r_off) * 3) % n) / (n - 1)
    g = low + span * (((i + g_off) * 5) % n) / (n - 1)
    b = low + span * (((i + b_off) * 7) % n) / (n - 1)
    return (r, g, b)
```

```python
# Set node and edge communities
set_node_community(G2, communities)
set_edge_community(G2)
node_color = [get_color(G2.nodes[v]['community']) for v in G2.nodes]
# Set community color for edges between members of the same community (internal) and intra-community edges (external)
external = [(v, w) for v, w in G2.edges if G2.edges[v, w]['community'] == 0]
internal = [(v, w) for v, w in G2.edges if G2.edges[v, w]['community'] > 0]
internal_color = ['black' for e in internal]
```

```python
pos = nx.spring_layout(G2, k=0.1)
plt.rcParams.update({'figure.figsize': (15, 10)})
nx.draw_networkx(
    G2,
    pos=pos,
    node_size=0,
    edge_color="#444444",
    alpha=0.05,
    with_labels=False)
```

```python
G4 = nx.read_edgelist('./compiledEdgelist.csv', comments='#',
                     create_using=nx.Graph(), 
                     delimiter=',', 
                     nodetype=str, 
                     encoding='utf-8')
print(G4)
```

```python
import csv

artistKeys = []

with open('artistKeys2.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        artistKeys.append(row[0])
print(artistKeys)
```

```python
#Kevin Bacon finder
noEdges = []
noPath = []
bacons = []

#processes all artists and checks against edgelist to determine if there exists a path between 2 nodes and the path length
for node in artistKeys:
    node_paths = []
    for node2 in artistKeys:
        if node != node2:
            if G4.has_node(node):
                if G4.has_node(node2):
                    if nx.has_path(G4, node, node2):
                        pathCount = nx.shortest_path_length(G4, node, node2)
    #                     print(pathCount)
                        node_paths.append(pathCount)
                    else:
#                         print("No path between: " + node + " and " + node2)
                        noPath.append([node, node2])
                else:
                    noEdges.append(node2)

#checks if all the path lengths for this node to every other reachable node is less than or equal to 6 aka is a Kevin Bacon
    if all(i <= 6 for i in node_paths):
#         print("FOUND A BACON\n")
        print(node)
        bacons.append(node)
print("Done")

noEdges = set(noEdges)
```

```python
#Writes all data to csv's
with open('noPathBetweenArtists.csv', 'w') as f:
    write = csv.writer(f)
    write.writerows(noPath)
with open('noEdges.csv', 'w') as f:
    write = csv.writer(f)
    write.writerows(noEdges)
with open('baconArtists.csv', 'w') as f:
    write = csv.writer(f)
    write.writerows(bacons)
```
